### Creating a function which plot 6 images from the dataset.

```python
def show(images):
    x = int(len(images)/5)
    
    plt.figure(figsize=(10,5))
    for i in range(6):
        plt.subplot(2,3,i+1)
        plt.imshow(images[i*x])
        
show(images)
```

### Resizing

```python
resized_images=[]
for i in range(len(images)):
    
    resize = tf.image.resize(images[i],[200,200])
    resized_images.append(resize)
    
for i in range(1):
    print("This is the Shape of resized image",resized_images[i].shape)
```

### Convert the colorspace of the image ( RGB to grey, grey to RGB, RGB to HSV, HSV to RGB, etc.)

```python
RGB_Grey = []
Grey_RGB = []
RGB_HSV = []
HSV_RGB = []

ColorCoversion={
"RGB_Grey" : [],
"Grey_RGB" : [],
"RGB_HSV" : [],
"HSV_RGB" : []    
}

for i in range(len(images)):
    converted1 = tf.image.rgb_to_grayscale(images[i])
    ColorCoversion["RGB_Grey"].append(converted1)

for i in range(len(RGB_Grey)):
    converted2 = tf.image.grayscale_to_rgb(RGB_Grey[i])
    ColorCoversion["Grey_RGB"].append(converted2)
    
for i in range(len(images)):
    converted3 = tf.image.rgb_to_hsv(images[i]/255)
    ColorCoversion["RGB_HSV"].append(converted3)
    
for i in range(len(RGB_HSV)):
    converted4 = tf.image.hsv_to_rgb(RGB_HSV[i])
    ColorCoversion["HSV_RGB"].append(converted4)
```

### Perform image adjustment on a given dataset ( Adjust Brightness, Contrast, Gamma, etc.)

```python
Brightness = []
Contrast = []
Gamma = []

for i in range(len(images)):
    converted1 = tf.image.adjust_brightness(images[i], delta=0.7)
    Brightness.append(converted1)
    
    converted2 = tf.image.adjust_contrast(images[i], 10)
    Contrast.append(converted2)
    
    converted3 = tf.image.adjust_gamma(images[i], 0.5)
    Gamma.append(converted3)
```

```python
show(Brightness)
```

```python
show(Contrast)
```

```python
show(Gamma)
```

### Perform Cropping, Flipping, Rotating, and Transposing through TensorFlow.

```python
crop = []
flip1 = []
flip2 = []
rot90 = []
transpose = []

for i in range(len(images)):
    converted1 = tf.image.central_crop(images[i], 0.5)
    crop.append(converted1)
    
for i in range(len(images)):
    converted2 = tf.image.flip_left_right(images[i])
    converted3 = tf.image.flip_up_down(images[i])
    flip1.append(converted2)
    flip2.append(converted3)
    
for i in range(len(images)):
    converted4 = tf.image.rot90(images[i])
    rot90.append(converted4)
    
for i in range(len(images)):
    converted5 = tf.image.transpose(images[i])
    transpose.append(converted5)
```
